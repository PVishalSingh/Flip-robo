```python
# Lets first connect to the web driver

driver=webdriver.Chrome(r"C:\Users\Visual singh\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
# to open flipkart for SAMSUNG Galaxy F13 (Waterfall Blue, 64 GB)

driver.get('https://www.flipkart.com/samsung-galaxy-f13-waterfall-blue-64-gb/p/itm583ef432b2b0c?pid=MOBGENJWBPFYJSFT&lid=LSTMOBGENJWBPFYJSFT1ZY7B0&marketplace=FLIPKART&q=smartphone&store=tyy%2F4io&srno=s_1_1&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&iid=2c5ce2f1-4a43-4a30-9152-2e17b94c2965.MOBGENJWBPFYJSFT.SEARCH&ssid=84ai51mke80000001667188856571&qH=1036ba3c4ed2c021')
```

```python
# To click all review option

All_Reviews=driver.find_element(By.XPATH,"//div[@class='_3UAT2v _16PBlm']")
All_Reviews.click()
```

```python
# Creating Empty List

urls=[]
short_review=[]
complete_review=[]
Rating=[]
time.sleep(2)
```

```python
start=0
end=400
for page in range(start,end):#for loop for scrapping 300 page
    
    short_reviews=driver.find_elements(By.CLASS_NAME,'_2-N8zT')#scraping review title 
    for i in short_reviews:
        short_review.append(i.text)
    
    complete_reviews=driver.find_elements(By.XPATH,"//div[@class='t-ZTKy']")#scraping detailed review
    for i in complete_reviews:
        complete_review.append(i.text)
    
    stars=driver.find_elements(By.XPATH,"//div[@class='_3LWZlK _1BLPMq']")#scraping detailed review
    for i in stars:
        Rating.append(i.text)
    
    nxt_button=driver.find_elements(By.XPATH,"//a[@class='_1LKTO3']")#scraping the list of buttons from the page
```

```python
print(len(Rating),len(short_review),len(complete_review))
```
